```python
import pandas as pd
import barcode
from barcode import Code128
from barcode.writer import ImageWriter

# Function to generate barcodes
def generate_barcode(data):
    try:
        barcode_value = Code128(str(data), writer=ImageWriter(), add_checksum=False)
        filename = f'barcode_{data}.png'
        barcode_value.save(filename)
        return filename
    except Exception as e:
        print(f"Error generating barcode for '{data}': {e}")
        return None

# Create a new column 'Barcode' and populate it with generated barcode values
df3['Barcode'] = df3['Product'].apply(generate_barcode)
```
